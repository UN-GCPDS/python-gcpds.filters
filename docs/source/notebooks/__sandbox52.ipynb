```python
from gcpds.filters import frequency as flt
import numpy as np
from matplotlib import pyplot as plt
from scipy.fftpack import rfft, rfftfreq
```

```python
t = np.arange(0, 10, 1/2500)
data = np.array([np.sin(2*np.pi*f*t) for f in [1, 10, 100, 1000]]).sum(axis=0)
```

```python
X = np.abs(rfft(data))
W = rfftfreq(X.size, t[1])
plt.semilogx(W, X)
```

```python
low100 = flt.GenericButterLowPass(f0=50, N=7)
dataf = low100(data, fs=2500);

X = np.abs(rfft(dataf))
W = rfftfreq(X.size, t[1])
plt.semilogx(W, X)
```

```python
low100 = flt.GenericButterHighPass(f0=50, N=7)
dataf = low100(data, fs=2500);

X = np.abs(rfft(dataf))
W = rfftfreq(X.size, t[1])
plt.semilogx(W, X)
```
